```python
class Student(object):
    
    def __init__(self, name, score):
        self.__name  = name   # 私有变量
        self.__score = score  # 私有变量
    
    def print_score(self):
        print('%s: %d' % (self.__name, self.__score))
        
    def get_name(self):
        return self.__name

    def get_score(self):
        return self.__score
    
    def set_score(self, score):
        if 0 <= score <= 100:
            self.__score = score
        else:
            raise ValueError('bad score')
```

```python
class Animal(object):
    def run(self):
        print('Animal is running...')
```

```python
class Dog(Animal):
    pass

class Cat(Animal):
    pass
```

```python
class Dog(Animal):

    def run(self):
        print('Dog is running...')

    def eat(self):
        print('Eating meat...')
```

```python
class Cat(Animal):

    def run(self):
        print('Cat is running...')

    def eat(self):
        print('Eating meat...')
```

```python
dog = Dog()
dog.run()
```
